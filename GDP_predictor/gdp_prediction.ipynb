```python
df['householdexpenditurepercapita'].fillna(method = 'ffill', inplace=True)
df['householdincomepercapita'].fillna(method = 'ffill', inplace=True)
df['consumerpriceindexgrowth'].fillna(method = 'ffill', inplace=True)
df["nonperformingloans(1k's)"].fillna(method = 'ffill', inplace=True)
df["NaturalGasImports(inK's)"].fillna(method = 'ffill', inplace=True)
df["crudeoilimports(100's)"].fillna(method = 'ffill', inplace=True)
df['subscribermobile'].fillna(method = 'ffill', inplace=True)
df['teledensitymobile'].fillna(method = 'ffill', inplace=True)
df['currencyissued(coins:INR1(inbn))'].fillna(method = 'ffill', inplace=True)
```

```python
df = df.set_index('year')
df_c = df.copy()
```

```python
from sklearn.impute import SimpleImputer 
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
df_imputed= imputer.fit_transform(df)
```

```python
x = DataFrame(df_imputed)
x
```

```python
x.index = df.index
x.columns = df.columns
y.index = df.index
```

```python
# Drop the columns with statistical significance (p-value)
# is greater than around 0.15 in successive backward eliminations.
import statsmodels.api as sm
X = np.append (arr=np.ones((51,1)).astype(float),values = x, axis =1)
X_opt = X[:,0:47] 
reg_ols = sm.OLS(endog=y, exog= X_opt).fit()
reg_ols.summary()
X_opt = DataFrame(X_opt)
X_opt = X[:,~X_opt.columns.isin([9,18,23,44])] 
reg_ols = sm.OLS(endog=y, exog= X_opt).fit()
reg_ols.summary()
X_opt2= DataFrame(X_opt)
X_opt2 = X_opt[:,~X_opt2.columns.isin([2,36])] 
reg_ols = sm.OLS(endog=y, exog= X_opt2).fit()
reg_ols.summary()
X_opt3= DataFrame(X_opt2)
X_opt3 = X_opt2[:,~X_opt3.columns.isin([13,17,18,26,27,30])] 
reg_ols = sm.OLS(endog=y, exog= X_opt3).fit()
reg_ols.summary()
X_opt4= DataFrame(X_opt3)
X_opt4 = X_opt3[:,~X_opt4.columns.isin([11,19,21])] 
reg_ols = sm.OLS(endog=y, exog= X_opt4).fit()
reg_ols.summary()
X_opt5= DataFrame(X_opt4)
X_opt5 = X_opt4[:,~X_opt5.columns.isin([19])] 
reg_ols = sm.OLS(endog=y, exog= X_opt5).fit()
reg_ols.summary()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_opt5,y,test_size=0.2,random_state=4)
```

```python
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

```python
# using linear regression

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
#polyreg = PolynomialFeatures(degree =3)
#x_poly = polyreg.fit_transform(x)
reg = LinearRegression()
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
#y_pred.index = y_test.index
y_pred = Series(y_pred)
y_pred.index = y_test.index
y_pred
```
